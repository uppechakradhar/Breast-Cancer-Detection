```python
#importing Libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('breast_cancer_wisconsin_cleaned.csv')
```

```python
df
```

```python
df.info()
```

```python
df['diagnosis'].value_counts()
```

```python
#Feature selection
from sklearn.model_selection import train_test_split
```

```python
diag_map = {'M':1,'B':0}
```

```python
df['diagnosis'] = df['diagnosis'].map(diag_map)
```

```python
df
```

```python
 x = df[['mean radius','mean perimeter','mean area','mean concavity','mean concave points']]
```

```python
x
```

```python
new_df = df.drop('diagnosis', axis=1, errors='ignore')
```

```python
new_df
```

```python
y = df[['diagnosis']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_df,y,test_size = 0.2, random_state=42)
```

```python
#model-KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn_y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(knn_y_pred,y_test)
```

```python
#model-Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state = 0)
```

```python
lr.fit(x_train,y_train)
```

```python
lr_y_pred = lr.predict(x_test)
```

```python
accuracy_score(lr_y_pred,y_test)
```

```python
#MODEL-NAIVE BAYES
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb_y_pred = gnb.predict(x_test)
```

```python
accuracy_score(gnb_y_pred,y_test)
```

```python
scores = cross_val_score(knn,x,y,cv = 10)
accuracy_all.append(accuracy_score(knn_y_pred,y_test))
accuracy_all.append(np.mean(scores))
print("Accuracy:{0:2%}".format(accuracy_score(knn_y_pred,y_test)))
print("cross validation score:{0:2%} (+/- {1:2%})".format(np.mean(scores),np.std(scores)*2))
```
